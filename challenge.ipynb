# Task 1- Load the Social Security data about baby names

```python
#imports

import numpy as np
import pandas as pd
```

```python
# initialize variables

names = None
names_year_list = [] 
```

```python
years = range(1880, 2018)
cols = ['Name', 'Sex', 'Number']
```

```python
# loop through the files to load dataset
for year in years:
    names_year = pd.read_csv('names/yob%d.txt' % year, names=cols)
    names_year['Year'] = year
    names_year_list.append(names_year)
    
# concatenate the dataframe one dataframe
names = pd.concat(names_year_list, ignore_index=True)
names.head()
```

# Task 2 â€“ Compare popularity trend of following names in form of a nice plot accross all the years since 1880
   a. ['Michael', 'John', 'David', 'Martin']
   
   b. ['Emily', 'Anna', 'Claire', 'Elizabeth']

```python
#index the data on gender, then name and then the year

names_sorted = names.set_index(['Sex','Name','Year']).sort_index(level=0)
names_sorted.head(20)
```

```python
import matplotlib.pyplot as plt

#A function which plots the values of this subset of data 
def draw(sex,name):
    sorted_subset = names_sorted.loc[sex,name]
    plt.plot(sorted_subset.index,sorted_subset.values)
```

#  Plot

```python
#Select some names for plotting.
selected_names = ['Michael','John','David','Martin']

#Select the size of the plot
plt.figure(figsize=(20,5))

#Call the 'draw' function for each name in above list.
for name in selected_names:
    draw('M',name)
plt.title("Matplotlib popularity: A scatter")  
plt.xlabel("Year")
plt.ylabel("Number")
plt.legend(selected_names)
plt.show()
```

# Extra Task - top 5 names in each year from 2001 to 2010 into single table

```python
#index the data on gender first, then name and then the year

names_sorted = names.set_index(['Sex','Name','Year']).sort_index(level=0)
names_sorted.head(20)

names_sorted.loc['M',:,2008].sort_values('Number',ascending=False).head()
```

```python
#Create a new DataFrame.
yob2008 = names_sorted.loc['M',:,2008].sort_values('Number',ascending=False)

#Use 'drop' method to delete columns and 'reset_index' to reset the index.
yob2008.reset_index().drop(['Sex','Year','Number'],axis=1).head()
```

```python
#Function yearwise to sort, drop and reset the index for single year.
def yearwise(sex,year):
    space = names_sorted.loc[sex,:,year].sort_values('Number',ascending=False).reset_index()
    space = space.drop(['Sex','Year','Number'],axis=1).head(5)
    space.columns = [year]
    space.index = space.index + 1
    return space

#Function sortring to call the 'yearwise_sorting'.
def sortring(sex,y1,y2):
    years = [yearwise(sex,year) for year in range(y1,y2+1)]
    return years[0].join(years[1:])

#calling function for 2001 to 2010.
print(sortring('M',2001,2010))
```

```python
#printing top 5  girls names in each year from 2001 to 2010.
print(sortring('F',2001,2010))
```
